```python
import pandas as pd
from os import listdir
from os.path import join
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
```

```python
# Settings
exploration = True
```

```python
data_path = "../data/"
fitbit_folder = "fitbit"

# Listing all the fitbit data folders
fitbit_path = join(data_path, fitbit_folder)
fitbit_folders = listdir(fitbit_path)
fitbit_folders
```

```python
# Let's create a dictionary to save the columns that contain a date
dates_in_data = {folder: [] for folder in fitbit_folders}
dates_in_data['heart-rate'] = ['Timestamp']
```

```python
# Let's print all markdown files in the data
def get_nested_markdown_files(folder_path, current_folder = ""):
    for path in listdir(folder_path):
        if path[-2:] == 'md':
            display(Markdown(f'## {current_folder}'))
            display(Markdown(open(join(folder_path, path), 'r').read()))
            continue
        if path[-2:] == 'gz':
            continue
        if path[-3:] == 'csv':
            continue
        get_nested_markdown_files(join(folder_path, path), path)
    return

get_nested_markdown_files(data_path)
```

```python
# Let's extract the first .gz file from each folder into a dataframe and make a list of them 
# Also, let's print out the readme of each of the data folders to know what we work with
data_frames = {}
for folder in fitbit_folders:
    folder_path = join(data_path, fitbit_folder, folder)
    folder_files = listdir(folder_path)
    for file_path in folder_files:
        if file_path[-2:] == 'gz':
            heart_path = file_path
        if file_path[-2:] == 'md':
            markdown_path = file_path
    f = open(join(data_path, fitbit_folder, folder, markdown_path), 'r') 
    display(Markdown(f'## {folder}'))
    display(Markdown(f.read()))
    data_frames[folder] = pd.read_csv(join(data_path, fitbit_folder, folder, heart_path), compression='gzip', decimal='.', parse_dates=dates_in_data[folder])
```

```python
# Let's also describe the data to get a feel for it (just the first file)
for (folder, df) in data_frames.items():
    display(Markdown(f'## {folder}'))
    display(df.describe())
```

```python
data_frames['heart-rate']['duration'] = data_frames['heart-rate']['Timestamp'].diff()
```

```python
max_time_between_measurements = pd.Timedelta(5, "m")
```

```python
heart_frame = data_frames['heart-rate'][data_frames['heart-rate']['duration'] <= max_time_between_measurements]
heart_frame = heart_frame.dropna()
```

```python
heart_frame['time_stamp'] = heart_frame['Timestamp'].apply(lambda x: x.ceil(freq='5T'))
heart_frame['time'] = heart_frame.time_stamp.dt.time
```

```python
heart_frame_5_minutes = heart_frame.groupby('time')['HeartRatePPG'].mean().reset_index(name='mean_heart_rate')
heart_frame_5_minutes['std'] = heart_frame.groupby('time').std(numeric_only=True).reset_index()['HeartRatePPG']
heart_frame_5_minutes['lower_bound90'] = heart_frame_5_minutes.mean_heart_rate - (1.645 * heart_frame_5_minutes['std'])
heart_frame_5_minutes['upper_bound90'] = heart_frame_5_minutes.mean_heart_rate + (1.645 * heart_frame_5_minutes['std'])

heart_frame_5_minutes['lower_bound95'] = heart_frame_5_minutes.mean_heart_rate - (1.96 * heart_frame_5_minutes['std'])
heart_frame_5_minutes['upper_bound95'] = heart_frame_5_minutes.mean_heart_rate + (1.96 * heart_frame_5_minutes['std'])
heart_frame_5_minutes
```

```python
# Let's plot some variables
heart_frame_5_minutes.plot('time',['lower_bound90', 'lower_bound95', 'mean_heart_rate', 'upper_bound90', 'upper_bound95'],figsize=(20,5))
```
